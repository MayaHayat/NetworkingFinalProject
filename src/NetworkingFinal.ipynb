```python
import pandas as pd
import matplotlib.pyplot as plt

def display(file_name):
    # Load the data from the CSV file
    data = pd.read_csv(file_name)

    # Calculate the inter-message delays
    data['Delay'] = data['Time'].diff()

    # Plot the inter-message delays as a line plot
    plt.plot(data['Delay'])
    plt.title('Inter-Message Delays')
    plt.xlabel('Message Number')
    plt.ylabel('Delay (s)')
    plt.grid(True)
    plt.show()

    # Plot the message sizes as a line plot
    plt.plot(data['Length'])
    plt.title('Message Sizes')
    plt.xlabel('Message Number')
    plt.ylabel('Size (bytes)')
    plt.grid(True)
    plt.show()

    # Plot the packet length as a function of time as a line plot
    plt.bar(data['Time'], data['Length'])
    plt.title('Packet Length as a Function of Time')
    plt.xlabel('Time (s)')
    plt.ylabel('Size (bytes)')
    plt.grid(True)
    plt.show()
    
        # Plot the PDF as a function of inter-message delays in orange
    plt.hist(data['Delay'].dropna(), bins=50, density=True, color='orange', alpha=0.7)
    plt.title('PDF of Inter-Message Delays')
    plt.xlabel('Inter-Message Delay (s)')
    plt.ylabel('Probability Density')
    plt.show()
    
```
